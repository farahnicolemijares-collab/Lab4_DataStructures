```python
from collections import defaultdict, Counter

# 1. Setup Unique Variables
CONTROL_NUM = 9
STUDENT_MAJOR = "BSECE"

# 2. Initialize the stream of Tuples
stream = [
    ("TUP Manila", "BSECE"),
    ("TUP Taguig", f"BSME_v{CONTROL_NUM}"),
    ("TUP Manila", f"BSME_v{CONTROL_NUM}"),
    ("TUP Manila", "BSECE"),
    ("TUP Visayas", STUDENT_MAJOR),
    ("TUP Taguig", "BSECE"),
    ("TUP Manila", "BSECE")
]

# 3. Data Expansion: Add (CONTROL_NUM + 3) = 12 custom tuples
for i in range(12):
    stream.append(("TUP Manila", "BSECE"))

# 4. Algorithm Phase 1 (Aggregation)
# Group programs by Campus
campus_map = defaultdict(list)
for campus, program in stream:
    campus_map[campus].append(program)

# 5. Algorithm Phase 2 (Frequency Analysis)
# Map programs to their frequencies using Counter
campus_counts = {}
for campus, programs in campus_map.items():
    campus_counts[campus] = Counter(programs)

# 6. Extract Top Requested Program for Manila
manila_top = campus_counts["TUP Manila"].most_common(1)[0]

print("--- Exercise 1: Admissions Results ---")
print(f"Total Applications Processed: {len(stream)}")
print(f"Top Program at TUP Manila: {manila_top[0]}")
print(f"Frequency: {manila_top[1]}")
```

```python
# Exercise 2
# 1. Setup
CONTROL_NUM = 9
FAVORITE_ARTIST = "Eraserheads"

# 2. Initialize Sets
festival = {"Ben&Ben", "SB19", "Bini", "Eraserheads", FAVORITE_ARTIST, "Zild", f"Indie Artist {CONTROL_NUM}"}

user_a = {"Ben&Ben", "Bini", "Maki", "Dionela", FAVORITE_ARTIST}
user_b = {"SB19", "Eraserheads", "Zild", f"Indie Artist {CONTROL_NUM}", "Parokya ni Edgar"}

# 3. Data Expansion: Add 9 new artists to the festival
festival.update({"Rivermaya", "Juan Karlos", "Lola Amour", "Dilaw", "Cup of Joe", "Munimuni", "Autotelic", "Any Name's Okay", "Over October"})

# 4. Algorithm: Jaccard Similarity J(A,B) = |A∩B| / |A∪B|
def calculate_jaccard(user_set, fest_set):
    intersection = user_set.intersection(fest_set)
    union = user_set.union(fest_set)
    return (len(intersection) / len(union)) * 100

sim_a = calculate_jaccard(user_a, festival)
sim_b = calculate_jaccard(user_b, festival)

# 5. Dealbreakers (Artists User A likes NOT in the festival)
dealbreakers = user_a.difference(festival)

print("\n--- Exercise 2: Music Recommendation ---")
print(f"User A Similarity: {sim_a:.2f}%")
print(f"User B Similarity: {sim_b:.2f}%")
print(f"User A Dealbreakers: {dealbreakers}")
```

```python
# Exercise 3
from collections import deque, Counter

# 1. Setup
CONTROL_NUM = 9
LOCAL_HAZARD = "FLOOD"

# 2. Initialize Deque with maxlen=5
burst_buffer = deque(maxlen=5)

# 3. Initial burst list
burst = [
    (CONTROL_NUM, "WEATHER"),
    (CONTROL_NUM + 1, "TRAFFIC"),
    (CONTROL_NUM + 2, "WEATHER"),
    (CONTROL_NUM + 3, "WEATHER"),
    (CONTROL_NUM + 4, LOCAL_HAZARD),
    (CONTROL_NUM + 5, "WEATHER"),
    (CONTROL_NUM + 6, "TRAFFIC")
]

# 4. Data Expansion: Add 11 custom alerts
for i in range(7, 18):
    burst.append((CONTROL_NUM + i, "WEATHER" if i % 2 == 0 else "TRAFFIC"))

# 5. Algorithm: Sliding Window with Spam Rule
for timestamp, category in burst:
    # Check current counts in buffer
    current_counts = Counter([item[1] for item in burst_buffer])
    
    # Spam Rule: If category exists >= 2 times, REJECT
    if current_counts[category] < 2:
        burst_buffer.append((timestamp, category))

print("\n--- Exercise 3: Sliding Window Results ---")
print(f"Final Buffer State: {list(burst_buffer)}")
print(f"Final Sum of Timestamps: {sum(item[0] for item in burst_buffer)}")
```
